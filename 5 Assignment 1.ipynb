```python
def FreqTable(ColName , dataset):
    FreqTable = pd.DataFrame(columns = ["Unique_Values" , "Frequency" , "Relative_Frequency", "Cumulative_Frequency"])
    FreqTable ["Unique_Values"] = dataset[ColName].value_counts().index
    FreqTable ["Frequency"] = dataset[ColName].value_counts().values
    FreqTable ["Relative_Frequency"] = (FreqTable ["Frequency"]/103)
    FreqTable ["Cumulative_Frequency"] = FreqTable ["Relative_Frequency"].cumsum()
    return FreqTable
```

```python
# Finding the outliers in the dataset
Table = pd.DataFrame (index = ["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%", "99%", "Q4:100%" , "IQR", "1.5rule", "LO", "GO", "Min", "Max"], columns = Quan) 
# IQR = Inter Quartile Range ; 1.5rule = Find 1.5 of IQR ; LO = Permissible Lesser Outlier Value ; GO = Permissible Greater Outlier Value ; Min = Actual Lesser Outlier Value; Max = Actual Greater Outlier Value
for ColName in Quan:
    Table [ColName] ["Mean"] = dataset [ColName].mean()
    Table [ColName] ["Median"] = dataset [ColName].median()
    Table [ColName] ["Mode"] = dataset [ColName].mode()[0]
    Table [ColName] ["Q1:25%"] = dataset.describe()[ColName]["25%"]
    Table [ColName] ["Q2:50%"] = dataset.describe()[ColName]["50%"]
    Table [ColName] ["Q3:75%"] = dataset.describe()[ColName]["75%"]
    Table [ColName] ["99%"] = np.percentile(dataset[ColName],99)
    Table [ColName] ["Q4:100%"] = dataset.describe()[ColName]["max"]
    Table [ColName] ["IQR"] = Table [ColName] ["Q3:75%"] - Table [ColName] ["Q1:25%"]
    Table [ColName] ["1.5rule"] =  Table [ColName] ["IQR"] * 1.5
    Table [ColName] ["LO"] =  Table [ColName] ["Q1:25%"] - Table [ColName] ["1.5rule"]
    Table [ColName] ["GO"] =  Table [ColName] ["Q3:75%"] + Table [ColName] ["1.5rule"]
    Table [ColName] ["Min"] = dataset [ColName].min()
    Table [ColName] ["Max"] = dataset [ColName].max()
```

```python
def SND(dataset):
    mean = dataset.mean()
    SD = dataset.std()
    Values = [i for i in dataset]
    Z_score = [((j-mean)/SD) for j in Values]
    sb.distplot(Z_score, kde = True, kde_kws = {'color' : 'blue'}, color = "Green")
    sum(Z_score) / len(Z_score)
```

```python
SND(dataset["salary"])
```

```python
def pdf_probability(dataset, startrange, endrange):
    from scipy.stats import norm
    Ax = sb.distplot(dataset, kde = True, kde_kws = {'color' : 'blue'}, color = "Green")
    plt.axvline(startrange, color = "Red")
    
    plt.axvline(endrange, color = "Red")
    
    # Generate sample of the dataset
    Sample = dataset
    
    # Calculate parameters
    Sample_mean = Sample.mean()
    Sample_SD = Sample.std()
    print("Mean = %.3f, Standard Deviation = %.3f" % (Sample_mean, Sample_SD))
          
    # Define the distribution
    Dist = norm(Sample_mean, Sample_SD) 

    # Sample probabilities for a range of outcomes
    Values = [Value for Value in range(startrange, endrange)]
    Probs = [Dist.pdf(Value) for Value in Values]
    Probability = sum(Probs)
    print ("The area between the range ({}, {}):{}". format (startrange, endrange, Probability))
    return Probability
```

```python
pdf_probability(dataset["salary"], 700000, 900000)
```
