# Codes to conduct ANOVA

```python
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("PrePlacement.csv")
dataset
```

```python
# Code to find if there is any null values in the dataset
dataset.isnull().sum()
```

```python
from scipy.stats import f_oneway
ANOVA = f_oneway(dataset['ssc_p'], dataset['hsc_p'], dataset['degree_p'])
ANOVA
```

```python
# The above codes are used to conduct ANOVA on the data set.
# In this case, a one-way ANOVA was conducted
# It is revealed that the p-value for the ANOVA is >0.05.
# This means the independent variables exhibit a significant difference.
# Because of this the end result is unreliable for the considered dataset.
```
